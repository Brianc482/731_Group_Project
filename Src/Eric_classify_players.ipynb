```python
#fromSaharsh
# Importing the label encoder
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df_new["encodedPosition"]= encoder.fit_transform(allData["Position"]) # Transforming the values into unique numbers from 1-n unique classes
positionT = encoder.inverse_transform([1]) # an example of position for an encoded value 
positionT
```

```python
#from here: https://towardsdatascience.com/pca-using-python-scikit-learn-e653f8989e60
#PCA using Python (scikit-learn)
from sklearn.preprocessing import StandardScaler
features = ['Overall', 'Potential', 'Value', 'International Reputation', 'Crossing',
       'Finishing', 'HeadingAccuracy', 'ShortPassing', 'Volleys', 'Dribbling',
       'Curve', 'FKAccuracy', 'LongPassing', 'BallControl', 'Acceleration',
       'SprintSpeed', 'Agility', 'Reactions', 'Balance', 'ShotPower',
       'Jumping', 'Stamina', 'Strength', 'LongShots', 'Aggression',
       'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure',
       'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving', 'GKHandling',
       'GKKicking', 'GKPositioning', 'GKReflexes']# Separating out the features
x = df_new.loc[:, features].values# Separating out the target
y = df_new.loc[:,['encodedPosition']].values# Standardizing the features
x = StandardScaler().fit_transform(x)
```

```python
x
```

```python
y
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])
```

```python
finalDf = pd.concat([principalDf, df_new[['encodedPosition']]], axis = 1)
```

```python
finalDf
```

```python
fig = plt.figure(figsize = (20,20))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = ['encodedPosition', 'Potential', 'Value']
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['encodedPosition'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1'], finalDf.loc[indicesToKeep, 'principal component 2'] , c = color , s = 50)
ax.legend(targets)
ax.grid()
```
