```python
# Importing libraries required for handling the data
import pandas as pd
import numpy as np

# Importing metrics library
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

# Importing libraries for preprocessing
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Importing library for splitting the data
from sklearn.model_selection import train_test_split

# Importing regression models from scikit-learn
from sklearn.linear_model import LinearRegression
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.ensemble import RandomForestRegressor
```

```python
reg = LinearRegression()
Germany = pd.read_csv("https://raw.githubusercontent.com/Brianc482/731_Group_Project/main/Data/Germany_cleaned.csv")
Argentina = pd.read_csv("https://raw.githubusercontent.com/Brianc482/731_Group_Project/main/Data/Argentina.csv")
Italy = pd.read_csv("https://raw.githubusercontent.com/Brianc482/731_Group_Project/main/Data/Italy_cleaned.csv")
Spain = pd.read_csv("https://raw.githubusercontent.com/Brianc482/731_Group_Project/main/Data/Spain_cleaned.csv")
Brazil = pd.read_csv("https://raw.githubusercontent.com/Brianc482/731_Group_Project/main/Data/Brazil.csv")
```

```python
target = "Overall"
predictors_GER = list(Germany.columns)
predictors_ARG = list(Argentina.columns)
predictors_ITA = list(Italy.columns)
predictors_SPA = list(Spain.columns)
predictors_BRA = list(Brazil.columns)


for r in [target,"Name","Age","Nationality","Position","Weak Foot","Preferred Foot", "Work Rate", "Release Clause"]:
  predictors_GER.remove(r)
  predictors_ARG.remove(r)
  predictors_ITA.remove(r)
  predictors_SPA.remove(r)
  predictors_BRA.remove(r)

print(target,predictors_GER)
```

```python
def numColumns(col):
    if type(list(Germany[col])[0]) == float:
        return col
numberPreds = []

for c in Germany.columns:
    temp = numColumns(c)
    if temp != None:
        numberPreds.append(temp)
        
scaler = StandardScaler()
numberPreds.append(target)
scaled_df = pd.DataFrame(data=scaler.fit_transform(Germany[numberPreds]),columns=numberPreds)
numberPreds.remove(target)

X = scaled_df[numberPreds]
y = scaled_df[target]

#print(X.shape,y.shape)
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
print("Germany regression models:\n")
print("MLP Regressor:")
mlpReg = MLPRegressor(max_iter=220)
mlpReg.fit(x_train.values, y_train.values)
preds = mlpReg.predict(x_test)
score = r2_score(y_test,preds)
mae_mlp = mean_absolute_error(y_test,preds)
#0.0 is best possible score from MAE
print ('Mean-Absolute-Error: ', mae_mlp)
#This is the weighted variance in our calculations 1.0 is maximum
print('R2 score:', score)

print("------------------------------------------------------")
print("Linear Regression:")
lr = LinearRegression()
lr.fit(x_train,y_train)
preds_lr = lr.predict(x_test)
score_lr = r2_score(y_test,preds_lr)
mae_lr = mean_absolute_error(y_test,preds_lr)
#0.0 is best possible score from MAE
print ('Mean-Absolute-Error: ', mae_lr)
#This is the weighted variance in our calculations 1.0 is maximum
print('R2 score:', score_lr)

print("------------------------------------------------------")
print("Random Forest Regressor:")
random = RandomForestRegressor(n_jobs=-1, n_estimators=100)
random.fit(x_train, y_train)
preds_rf = random.predict(x_test)
score_rf = r2_score(y_test,preds_rf)
mae_rf = mean_absolute_error(y_test,preds_rf)
#0.0 is best possible score from MAE
print ('Mean-Absolute-Error: ', mae_rf)
#This is the weighted variance in our calculations 1.0 is maximum
print ('R2 score: ', score_rf)
```
